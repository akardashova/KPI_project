```python
# Добавим праздники как отдельный признак
holidays_kz = set(holidays.KZ(years=(2020, 2021, 2022, 2023)).keys())
def is_holiday(date):
    if datetime.date(date) in holidays_kz:
        return 1
    return 0
    
df['is_holiday'] = df['event_date'].apply(is_holiday)
df
```

```python
# Посмотрим на ковид
covid = df[(df.year == 2020) & ((df.month == 3)|(df.month == 4)|(df.month == 5)|(df.month == 6))]
sns.set_theme(rc={'figure.figsize':(13,5)})
plt.xlabel('Дата')
plt.ylabel('Количество контактов')
sns.lineplot(x=covid.event_date, y=covid.y);
```

```python
# Погуглила даты ЧП с ограничением выхода в Казахстане: с 15 марта по 11 мая, отметим эти данные
def is_covid(date):
    if date > pd.to_datetime('2020-03-15') and date < pd.to_datetime('2020-05-11'):
        return 1
    return 0

# Также добавила признак "Новый год", потому что из всех праздников он самый особенный и у нас там сильнее, 
# чем в другие праздники проседает показатель
def is_NewYear(date):
    if (date.month == 1 and date.day == 1) or (date.month == 12 and date.day == 31):
        return 1
    return 0


df['is_covid'] = df['event_date'].apply(is_covid)
df['is_NewYear'] = df['event_date'].apply(is_NewYear)
df
```

```python
X = df[['year', 'month', 'day', 'week_day', 'is_covid', 'is_holiday', 'is_NewYear']]
# Сделаем признаки категориальными
X = pd.get_dummies(X, columns=['month', 'day', 'week_day'], drop_first=True)
X
```

```python
# Будем обучаться сначала на 2020 и 2021, потом будем валидироваться на 2022. Подберем параметры, после чего обучимся на всех данных и предскажем 2023
X_train = X[:365*2+2]
X_val = X[365*2+2:]
y_train = df.y[:365*2+2]
y_val = df.y[365*2+2:]
```

```python
X_test = pd.DataFrame(pd.date_range('2023-01-01', '2023-12-31'))
X_test['year'] = X_test[0].dt.year
X_test['month'] = X_test[0].dt.month
X_test['day'] = X_test[0].dt.day
X_test['week_day'] = X_test[0].dt.weekday
X_test['is_holiday'] = X_test[0].apply(is_holiday)
X_test['is_covid'] = X_test[0].apply(is_covid)
X_test['is_NewYear'] = X_test[0].apply(is_NewYear)
X_test = X_test[['year', 'month', 'day', 'week_day', 'is_covid', 'is_holiday', 'is_NewYear']]
X_test = pd.get_dummies(X_test, columns=['month', 'day', 'week_day'], drop_first=True)
X_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_transformed = scaler.transform(X_train)
X_val_transformed = scaler.transform(X_val)
X_test_transformed = scaler.transform(X_test)
X_train_transformed
```

```python
from LinearRegression import MyLineReg
```

```python
# Подберем параметры модели
linear_regression = MyLineReg()

def set_params(model, params):
    """
    Устанавливает параметры модели.
    
    Параметры:
        model (объект модели): Модель, параметры которой нужно установить.
        params (словарь): Словарь параметров, которые нужно установить.
    """
    for param, value in params.items():
        setattr(model, param, value)

def calculate_mape(y_true, y_pred):
    """
    Вычисление MAPE (Mean Absolute Percentage Error).
    """
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def parameter_search(param_grid, model, X_train, y_train, X_test, y_test):
    """
    Функция для перебора параметров модели и выбора параметров с наименьшим MAPE.
    """
    best_mape = float('inf')
    best_params = None

    prod = list(itertools.product(*param_grid.values()))
    for params in tqdm(prod, total = len(prod)):
        param_dict = dict(zip(param_grid.keys(), params))
        set_params(model, param_dict)
        model.fit(X_train, y_train, verbose=False)
        y_pred = model.predict(X_test)
        mape = calculate_mape(y_test, y_pred)
        if mape < best_mape:
            best_mape = mape
            best_params = param_dict

    return best_params, best_mape


params = {'n_iter': [1000, 5000],
          'learning_rate': [0.001, 0.005, 0.01],
          'reg': ['l1', 'l2', 'elsaticnet'], 
          'l1_coef': [0, 0.2, 0.5, 0.7], 
          'l2_coef': [0, 0.2, 0.5, 0.7]}

best_params, best_mape = parameter_search(params, linear_regression, X_train_transformed, y_train, X_val_transformed, y_val)

print("Лучшие параметры:", best_params)
print("Лучшее значение MAPE:", best_mape)
```

```python
linear_regression = MyLineReg(**best_params)
linear_regression.fit(X_train_transformed, y_train, verbose=False)
```

```python
# Смотрим как мы предиктим трейн
plt.plot(df['event_date'][:365*2+2], linear_regression.predict(X_train_transformed), label='Прогноз')
plt.plot(df['event_date'][:365*2+2], y_train, label='Реальные данные')
plt.legend();
```
